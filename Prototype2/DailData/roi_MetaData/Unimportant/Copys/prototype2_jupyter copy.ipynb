```python
# Imports
import pandas as pd
```

```python
# Check Columns in csv
csv_filename = "final_dataset_model.csv"

df = pd.read_csv(csv_filename)
print(df.columns.tolist())
```

```python
speakers_per_province = df.groupby("province")["speaker_key"].nunique().sort_values(ascending=False)

print("Speakers per province:")
print(speakers_per_province)
print("\nTotal unique speakers:", df["speaker_key"].nunique())
```
